## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
f = plt.figure()
a1 = f.add_axes([0, 0, 1, 1])
a2 = f.add_axes([0.2, 0.5, 0.2, 0.2])
a1.plot(0, 0,  color = 'red')
a2.plot(0, 0, color = 'red')
a1.set_xlim(0, 1)
a1.set_ylim(0, 1)
a2.set_xlim(0, 1)
a2.set_ylim(0, 1)
a2.set_xticks([0.2, 0.4, 0.6, 0.8, 1])
a2.set_yticks([0.2, 0.4, 0.6, 0.8, 1])
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
f = plt.figure()
a1 = f.add_axes([0, 0, 1, 1])
a2 = f.add_axes([0.2, 0.5, 0.2, 0.2])
a1.plot(x, y,  color = 'brown')
a2.plot(x, y, color = 'brown')
a2.set_xticks([20, 40, 60, 80, 100])
a2.set_yticks([50, 100, 150, 200])
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
f = plt.figure()
a1 = f.add_axes([0, 0, 1, 1])
a2 = f.add_axes([0.2, 0.5, 0.4, 0.4])
a1.plot(0, 0)
a2.plot(0, 0)
a1.set_xlim(0, 1)
a1.set_ylim(0, 1)
a2.set_xlim(0, 1)
a2.set_ylim(0, 1)
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
f = plt.figure()
a1 = f.add_axes([0, 0, 1, 1])
a2 = f.add_axes([0.2, 0.5, 0.2, 0.2])
a2.set_xlim(20,22)
a2.set_ylim(30, 50)
a1.plot(x,z, color = 'blue')
a2.plot(x,y, color = 'blue')

l = [20+.5*i for i in range(0,5)]
p = [30+5*i for i in range(0, 5)]
a2.set_xticks(l)
a2.set_title("zoom")
a2.set_yticks(p)
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
plt.subplots(nrows = 1, ncols = 2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
f, a = plt.subplots(nrows = 1, ncols = 2)
a[0].plot(x, y,ls = '--', color = 'blue')
a[1].plot(x, z, color = 'red')
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
f, a = plt.subplots(nrows = 1, ncols = 2, figsize=(12, 1.8))
a[0].plot(x, y,lw = 3, color = 'blue')
a[1].plot(x, z,lw = 3,ls = '--', color = 'red')
a[0].set_yticks([0,50,100,150,200])
a[1].set_yticks([0,2500,5000,7500,10000])
```
